## Coolest Discovery: Gemini Grading Consistency

### The Problem

Gemini was grading inconsistently - the same answer would get different scores on different attempts. Consistency was only around 60%.

### The Discovery

AI models are probabilistic, not deterministic. Even identical inputs can produce different outputs because of how language models generate text (they sample from probability distributions).

### The Solution

**Technical Changes:**
1. Set temperature parameter to 0 (most deterministic setting)
2. Use structured output format (JSON only)
3. Provide detailed rubrics in prompt
4. Include 3 example answers with scores in prompt

### The Result

This single discovery transformed our grading system from "interesting experiment" to "actually usable tool."

### Why This Matters

Understanding how models work lets you engineer better solutions.
