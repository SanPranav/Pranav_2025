## Main Project Highlight: Calendar

The strongest thread in my 2025-2026 work is the calendar system. It is not just a visual page. It became a shared coordination layer between the student-facing UI, the Spring backend, the email/comment pipeline, and the game flow. That is why it deserves the main spotlight.

### 1. The calendar starts as a real data model, not a mockup

The first major step was building the actual `CalendarIssue` backend instead of faking the page state. That work introduced the issue model, controller, repository, status, and priority types in [46c22dea](https://github.com/aadibhat09/spring/commit/46c22deaf53db66972ba260769adbb028e64f022). After that, I extended the behavior with requested features and comment support in [36bbf60f](https://github.com/aadibhat09/spring/commit/36bbf60fdbf3021113b9b9ed496e7ce0fa59c253) and then tightened the service/controller flow again in [1a06320e](https://github.com/aadibhat09/spring/commit/1a06320ef871bea34b8549b10b7a94eaa4809dd6) and [9ea6a037](https://github.com/aadibhat09/spring/commit/9ea6a037e7c6583c04fe55fb9fac229612bf20f6).

### 2. The UI evolved into an actual authored experience

The markdown preview and author attribution landed in [49f2bed6](https://github.com/aadibhat09/spring/commit/49f2bed6925e06bdb50ec7439ebb35595352484b), which made the calendar feel like an editorial workflow instead of a static list. On the pages side, the same theme kept moving forward through [c0539b2c](https://github.com/aadibhat09/pages-fork/commit/c0539b2c062adcdaed9b9c6ce13c052f753a9463), [ea1763fa](https://github.com/aadibhat09/pages-fork/commit/ea1763fac56097df413b19a1176102493fdf755f), [ffd236ce](https://github.com/aadibhat09/pages-fork/commit/ffd236ce126b8c68d47ca4768625c7a9d194fad7), [54a1caf5](https://github.com/aadibhat09/pages-fork/commit/54a1caf5a90d5559a8b68df094ec743648d08c3e), and [b1bc51bb](https://github.com/aadibhat09/pages-fork/commit/b1bc51bb741d26b32637cf3192bde83c131db472).

### 3. Styling was treated as a system, not a decoration pass

I pushed the calendar styling into `_sass/open-coding/calendar.scss` and then kept tightening it through [8232a3c2](https://github.com/aadibhat09/pages-fork/commit/8232a3c2c5fec1e9af6152f017c848e0898b58d7), [506442114](https://github.com/aadibhat09/pages-fork/commit/50644211439ce2dff17a849731f0b224c01fb808), [29caaceb](https://github.com/aadibhat09/pages-fork/commit/29caaceb91d78f7a7ba1354d922f3779d2ffb3a0), [a6be1e64](https://github.com/aadibhat09/pages-fork/commit/a6be1e6486c03cccd0b93f3907b2fb389383310d), and [af0dea2de](https://github.com/aadibhat09/pages-fork/commit/af0dea2deef4f998220fa3f49e233ea0e329a84c). The later cleanup in [54a1caf5](https://github.com/aadibhat09/pages-fork/commit/54a1caf5a90d5559a8b68df094ec743648d08c3e) and [b1bc51bb](https://github.com/aadibhat09/pages-fork/commit/b1bc51bb741d26b32637cf3192bde83c131db472) helped keep the page readable as the feature set grew.

### 4. The backend moved from issue tracking into notifications and collaboration

Once the calendar existed, I extended it with comments and email-driven interactions. The important backend jump is [6b248322](https://github.com/aadibhat09/spring/commit/6b2483227ed70af106721b5f7240346ed45e0098), which added the comment API, email service changes, and test coverage. That work was stabilized in [a71bb3aa](https://github.com/aadibhat09/spring/commit/a71bb3aa5f09b4db9782eccf2e26cf579825e1ed), [857bb2a1](https://github.com/aadibhat09/spring/commit/857bb2a16d6be8c52037999c2d03549ba31a43c7), [d0d5cdee](https://github.com/aadibhat09/spring/commit/d0d5cdee764acad682cd161f1a978ae70f039ce8), and [33c47dea](https://github.com/aadibhat09/spring/commit/33c47deac331b95a49b4b0011fda2813d94817be).

### 5. The calendar finally connected to the game layer

The most distinctive part of the project is the bridge from the calendar into the game and pathway systems. That bridge appears in [8b64ce4ac](https://github.com/aadibhat09/pages-fork/commit/8b64ce4ac5e020141ef0866fa120fecaf5219997), which tried the calendar-to-game link, and then in [8bcb15a96](https://github.com/aadibhat09/pages-fork/commit/8bcb15a96d530fcbb34168e9e0ac319ad11d890c), which fully integrated the personal calendar timeline with the student calendar UI. The supporting path continues through `PathwayCalendarApi.js`, `PathwayCalendarCoordinator.js`, `PathwayTimelineBuilder.js`, and `GameLevelCsPath0Forge.js` in that commit.

### 6. The result is a real portfolio system, not a single feature

The calendar work now spans content, appearance, backend behavior, and game-based presentation. That is why it is my main highlight. The commit trail shows the progression clearly: model first, then authoring, then styling, then notifications, then game integration.
