## Before the Event

I was excited beacuse I wanted to present for both Optix and CSA, for which I both as able to do. I expected other robotics teams which I had friends in.

## Curiosity Before Attending

Other teams and how they functioned as well as the other projects that CSP has done over the year.

## Most Interesting Thing I Saw

Definitley the automobile car that could drive around that they built as a class in poway high school.

## What I Learned

One specific thing I learned was how different engineering and computer science programs approach teamwork and specialization. Each robotics team had different systems for CAD, programming, electrical, and outreach, which showed me how important organization and communication are in large technical projects.

## What Surprised Me

I was surprised by how advanced some of the student-built projects were, especially the automobile project from Poway High School and the variety of real-world applications students were able to demonstrate through robotics, programming, and engineering.

## What Inspired Me

What inspired me the most was seeing how much effort students put into creating projects that were both technically impressive and genuinely useful. Watching teams confidently explain their designs and programming showed how much growth can happen through hands-on experience.

## Personal Connection

This connected to my personal interests because I enjoy computer science, robotics, and building projects that combine software and hardware. Presenting NodCursor and representing Optix also made me reflect on how much I have improved in communication, programming, and teamwork throughout the year.

## Interaction

**Person or group I learned from:** Presenter/Student

They explained how their projects were developed, how their teams organize responsibilities, and the technical challenges they faced during the build process. The automotive students showed how their vehicle was assembled and modified, while robotics teams explained programming logic, mechanisms, and strategy.

## Depth of Experience

This event showed me how broad the field of engineering and computer science really is. It helped me realize that there are many different ways technology can be applied creatively, whether through robotics, software, automotive engineering, or agriculture.

## Connection to CS or Future Goals

This connects directly to my CS class and projects because presenting NodCursor required many of the same skills we learn in class, including problem solving, collaboration, debugging, and explaining technical ideas clearly. It also reinforced my interest in pursuing computer science and engineering-related projects in the future.

## Final Reflection

This event was meaningful to me because I was able to both present my own work and learn from other students across different technical pathways. It gave me a better understanding of how much innovation students are capable of when they work on projects they are passionate about.

## Biggest Takeaway

> My biggest takeaway was that collaboration and hands-on experience are some of the most important parts of learning engineering and computer science.
