* **Date:** May 21
* **Location:** A101
* **Theme:** What You Saw and What You Learned

---

## Before the Event

* Expected a standard entrepreneurship presentation focused on:

  * Building a company
  * Funding strategies
  * Product development
* Assumed the talk would connect somewhat to computer science, but mostly at a surface level
* Thought the focus would be more technical than personal

---

## Curiosity Before Attending

* Wanted to learn how people without traditional technical backgrounds enter:

  * Tech
  * Civic innovation
  * Entrepreneurship
* Interested in understanding:

  * How ideas become real companies
  * The business and relationship side of innovation
  * What happens before a product is fully built

---

## What I Saw

* Presentation by Elijah Johnson, CEO of SafeBallot
* SafeBallot:

  * Blockchain-based digital voting platform
  * Focused on election security and civic engagement
* Event combined:

  * Entrepreneurship lessons
  * Civic technology
  * Personal experiences transitioning from D1 athletics to startup leadership
* Learned about the Playbook Project:

  * Community initiative connecting 18 cities
  * Helped provide food for 450 families across the county

---

## Most Interesting Thing

* SafeBallot's early attention came from:

  * Its branding
  * The company name
  * The logo
* Investors responded before the product was fully developed
* Changed my perspective on entrepreneurship:

  * Strong presentation and identity can create opportunities before technical completion

---

## What I Learned

* People often invest in:

  * The founder
  * The vision
  * The authenticity behind the product
* Technical quality alone is not enough
* Communication, trust, and personality matter just as much as engineering
* A strong emotional connection can determine whether people support an idea

---

## What Surprised Me

* Major opportunities came through relationships and consistency:

  * His wife's connection helped open a partnership opportunity in Mozambique
  * A conversation at a pitch competition eventually led to presenting before Congress
* Success was not always planned strategically
* Repeated exposure and genuine networking created momentum over time

---

## What Inspired Me

* The idea of:

  > "Start before you're ready."
* Elijah Johnson began pitching with:

  * A vision
  * A logo
  * A mission
* He did not wait for a perfect or finished product
* Inspired me to:

  * Take more initiative on projects
  * Be comfortable starting from an incomplete position
  * Trust growth through action and consistency

---

## Personal Connection

* In computer science, I usually focus heavily on:

  * Technical rigor
  * Optimization
  * Security
* The talk reminded me that:

  * Technical work only creates impact when people trust and understand it
  * Communication and relationships are essential parts of building meaningful systems
* Made me realize I should improve the way I:

  * Present ideas
  * Explain technical concepts
  * Build credibility with others

---

## Interaction

### Person I Learned From

* Elijah Johnson

  * CEO of SafeBallot
  * Former D1 football player
  * Entrepreneur and speaker

### Key Points He Shared

* Entered entrepreneurship competitions before being fully prepared
* Faced repeated rejection:

  * "99 no's before 1 yes"
* Used unconventional methods to create opportunities:

  * Sponsoring a basketball team
  * Picking up trash on a highway to get in front of a client
* Emphasized:

  * Consistency
  * Discipline
  * Authenticity
  * Persistence over perfection

---

## Depth of Experience

### My Perspective Before

* Thought technical skill was the main barrier to creating impact

### My Perspective After

* Realized the larger barrier is often:

  * Human connection
  * Communication
  * Relationship-building
  * Consistency
* Technical systems matter, but people create opportunities

---

## Connection to CS and Future Goals

* Connected strongly to my work in Open Coding Society:

  * Backend systems
  * Security architecture
* Realized cybersecurity is not only about protection:

  * It is also about trust and credibility
* Important lessons:

  * Stakeholders must understand and trust the system
  * Security posture can influence partnerships and investment decisions
  * Communication is part of technical leadership
* Changed how I think about presenting technical work:

  * Not just execution
  * Also perception, trust, and confidence

---

## Final Reflection

* The event reframed what it means to be a builder
* I usually focus on optimizing:

  * Code
  * Systems
  * Technical solutions
* The talk emphasized optimizing:

  * Authenticity
  * Consistency
  * Human connection
* Showed how persistence and relationships can transform a small idea into something with national visibility

---

## Biggest Takeaway

> "No dream is too small to start, and no idea is too big to pursue - as long as you are willing to stay consistent, authentic, and resilient through rejection."
